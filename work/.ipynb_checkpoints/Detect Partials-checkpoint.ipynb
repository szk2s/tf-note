```python
import json
import numpy as np
import librosa.display
import librosa
from skimage.measure import label
from matplotlib import pyplot as plt
import plotly.offline as py
import plotly.graph_objs as go
```

```python
def locate(condition):
    i, j = np.where(condition)
    pos = np.empty((i.size + j.size,), dtype=i.dtype)
    pos[0::2] = i
    pos[1::2] = j
    pos = np.split(pos, i.size)
    pos = [tuple(item) for item in pos]
    return pos
```

```python
with open('./input/bird.json') as input_json:  
    input_obj = json.load(input_json)
    
songname = input_obj['songname']
S = np.array(input_obj['spectrogram'])
segment_idx = label(np.array(input_obj['segmentation'])) - 2
```

```python
times = librosa.frames_to_time(np.arange(S.shape[1]))
freqs = librosa.fft_frequencies()
time_grid, freq_grid = np.meshgrid(times, freqs)
```

```python
# Plot Spectrogram
plt.figure()
librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max), y_axis='log', x_axis='time')
plt.title('Power spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
```

```python
# Plot segments
plt.figure()
librosa.display.specshow(librosa.amplitude_to_db(segment_idx + 1,ref=np.max), y_axis='log', x_axis='time')
plt.title('Power spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
print(np.unique(segment_idx, return_counts=True))
```

```python
def find_partial_in_segment(S, segment):
    i, j = np.where(segment)
    segment_columns = np.unique(j)
    partial_positions = []
    for column_idx in segment_columns:
        amp_column = S[:, column_idx]
        is_segment = segment[:, column_idx]
        max_val = amp_column[is_segment].max()
        row_idx = np.where(amp_column == max_val)[0][0]
        partial_positions.append((row_idx, column_idx))
    return partial_positions
```

```python
partials_positions = []
partials = []
for i in range(segment_idx.max()):
    segment = segment_idx==i
    partial_positions = find_partial_in_segment(S, segment)
    partials_positions.append(partial_positions)
    partial = dict(id=i)
    partial['points'] = [dict(time=time_grid[position], 
                              freq=freq_grid[position], 
                              amp=S[position]) 
                         for position in partial_positions]
    partials.append(partial)
```

```python
# plot is_partial
is_partial = np.zeros(S.shape, dtype = int)
for partial_positions in partials_positions:
    for position in partial_positions:
        is_partial[position] += 1
print(is_partial.min(), is_partial.max())

plt.figure()
librosa.display.specshow(is_partial, y_axis='log', x_axis='time')
plt.title('Power spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
```

```python
with open('./output/bird.json') as original_json:  
    output_obj = json.load(original_json)
    output_obj['partials'] = partials

with open('./output/bird.json', 'w') as outfile:
    json.dump(output_obj, outfile)
```

```python
def create_trace(partial_positions):
    x_list = [times[position[1]] for position in partial_positions]
    y_list = [freqs[position[0]] for position in partial_positions]
    trace = go.Scatter(x = x_list, 
                       y = y_list, 
                       mode = 'lines+markers',
                       line = dict(width = 6),
                       marker= dict(size= 10))
    return trace

data = [create_trace(partial_positions) for partial_positions in partials_positions]

spectrogram = go.Heatmap(x=times,
                         y=freqs,
                         z=librosa.amplitude_to_db(S,ref=np.max))

data.append(spectrogram)

layout = go.Layout(
    title='Partial Data',
    height=800,
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    ),
    xaxis=dict(
        title='Time',
        ticklen=5,
        gridwidth=2,
    ),
    yaxis=dict(
        title='Frequency',
        type='log',
        ticklen=5,
        gridwidth=2,
    )
)

py.plot(go.Figure(data, layout), filename='./partial_data.html')
```
